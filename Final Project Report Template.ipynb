##### Optimization
1. Drop kindergarten and find out the key schools. This part is brought forward in the data cleaning part (3.2.4) because we want to keep the integrity of our code. 
2. Colinearity test. One of the main assumptions of multiple linear regression models is that our predictive variables are independent of each other. We want the predictive variables (independent variables) to be related to the response variables (dependent variables), not to each other. Variance inflation factor (VIF), is the ratio of the Variance between the explanatory variables with multicollinearity and the variance without multicollinearity. The larger the VIF, the more collinearity it shows. The empirical judgment method shows that when $0 < VIF < 10$, there is no multicollinearity. When $10 \le VIF \ge 100$, there is a strong multicollinearity. When $VIF \ge 100$, there is serious multicollinearity. We wrote a function to test the VIF. The code are as follows:

```python
def vif(df, col_i):
    cols = list(df.columns)
    cols.remove(col_i)
    cols_noti = cols
    formula = col_i + '~' + '+'.join(cols_noti)
    r2 = ols(formula, df).fit().rsquared
    return 1. / (1. - r2)

params = ['Area','Key_school','School', 'Bedroom', 'Living_room','East','South','North','West','City_North','City_South','City_West','Old','Middle_floor','Low_floor','Else','Roughcast','Simple_renovated']
test_data = result[params]
test_data
for i in test_data.columns:
    print(i, '\t', vif(df=test_data, col_i=i))
```

#### SVR
For svr model, we choose linear as its kernel. As a result, this model is also a linear regression model with a different way of calculating the errors. SVR acts as a potential subsitution of multi linear regression.
##### Preprocessing of data
To apply support vector regression, we have to preprocess our data. We first split the data into train data and test data, 80 percent for training and 20 precent for testing. Then, we standardize the data as a prepration for fitting them into the model. 

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error,mean_squared_error,explained_variance_score,r2_score
from sklearn.model_selection import train_test_split
import numpy as np
#加载数据集
x = result[['Area','Key_school','School', 'Bedroom', 'Living_room','East','South','North','West','City_North','City_South','City_West','Old','Middle_floor','Low_floor','Else','Roughcast','Simple_renovated']]
y = result['Price_10K']

# 拆分数据集
x_train,x_test1,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
# 预处理
y_train = np.array(y_train).reshape(-1, 1)
y_test = np.array(y_test).reshape(-1, 1)
x_train = StandardScaler().fit_transform(x_train)
x_test0 = StandardScaler().fit_transform(x_test1)
y_train = StandardScaler().fit_transform(y_train).ravel()
y_test = StandardScaler().fit_transform(y_test).ravel()
#创建svR实例
svr=SVR(C=1, kernel='linear')
svr=svr.fit(x_train,y_train)
#预测
svr_predict=svr.predict(x_test0)
```

##### Evaluation of model
After biulding the model, we managed to get the MAE, MSE, EVS, and R2, which are some measurements of the accuracy of the model. Particularly, variance for our model is about 0.776, which shows that the accuracy of our model is quite good but not as good as that of MLR's. 

```python
#评价结果

maerbf = mean_absolute_error(y_test, svr_predict)
mserbf = mean_squared_error(y_test, svr_predict)
evsrbf = explained_variance_score(y_test, svr_predict)
r2rbf = r2_score(y_test, svr_predict)
print("MAE：", maerbf)
print("MSE：", mserbf)
print("EVS：", evsrbf)
print("R2： ", r2rbf)
```

#### 3. Evaluation of the model (PCA)
In order to visaulize the data, we have to reduce our data into one dimension, and we need to rely on principle components analysis. We first calculate the variance ratio of each principle component. The variance ratio of the principle component is 0.998, which means that the first principle compinent contains 99.8 precent of the information edited before. We finally reduce data into one dimension with the help of these codes.  

```python
from sklearn.decomposition import PCA
X = result[params]
pca = PCA(n_components='mle')
pca.fit(X)
# PCA(copy=True, n_components=2, whiten=False)
print(pca.explained_variance_ratio_)
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

def pca(X,k):
    n_samples, n_features = X.shape
    mean = np.array([np.mean(X[:,i]) for i in range(n_features)])
    norm_X = X-mean
    scatter_matrix=np.dot(np.transpose(norm_X),norm_X)
    eig_val, eig_vec = np.linalg.eig(scatter_matrix)
    eig_pairs = [(np.abs(eig_val[i]), eig_vec[:,i]) for i in range(n_features)]
    eig_pairs.sort(reverse=True)
    feature=np.array([ele[1] for ele in eig_pairs[:k]])
    data=np.dot(norm_X,np.transpose(feature))
    return data

x = result[['Area','Key_school','School', 'Bedroom', 'Living_room','East','South','North','West','City_North','City_South','City_West','Old','Middle_floor','Low_floor','Else','Roughcast','Simple_renovated']]
y = result['Price_10K']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)

X_train = []
X_test = []
for i in range(0,len(x_train)):
    X = x_train[params].iloc[i].values.reshape(-1,1)
    X_train.append(pca(X,1)[0][0])
for i in range(0,len(x_test)):
    X = x_test[params].iloc[i].values.reshape(-1,1)
    X_test.append(pca(X,1)[0][0])
Y_train = y_train.values
result_train = result.iloc[x_train.index]
a = ols('Price_10K ~ Area + Key_school + School + Bedroom + Living_room + East + South + North + West + City_North + City_South + City_West + Old + Middle_floor + Low_floor + Else + Roughcast + Simple_renovated', data = result_train).fit()
Y_predict = pd.DataFrame(x_test*a.params[1:].astype(float)).sum(axis=1).values
```

### 1）The influence of district types on house prices

We can see from the data that house prices in the west of the city are the most expensive. This may be because the environment in the west of the city is better, and the transportation and public facilities are better. ### 2）The influence of floor types on house prices

From the data and multiple linear regression model we can see that people prefer to buy houses on lower floors. Houses with lower floors are more expensive than houses with medium floors, and houses with medium floors are more expensive than houses with high floors.
![](floor.png)

### 3）The influence of key schools on house prices

In China, every family attaches great importance to children's education. This emphasis is particularly reflected in K12 education, where every family wants to send their children to the best primary and secondary schools. As a result of China's education policy, every school has its own school district. Children in the school district can study in the school for free. As a result, key school districts will make housing prices rise.
Through the model analysis, we can find that the impact of key schools on housing prices is very high. The price of key schools’ district housing is more expensive.
